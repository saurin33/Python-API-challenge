```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import requests
import gmaps
from api_keys import g_key
from pprint import pprint
```

```python
cities_df = pd.read_csv('../output_data/cities.csv')
cities_df
```

```python
cities_df['Humidity'].dtype
```

```python
locations = cities_df[['Lat', 'Lng']]
humidity = cities_df['Humidity']
```

```python
gmaps.configure(api_key = g_key)
```

```python
data = cities_df[['Lat', 'Lng', 'Humidity']]
data.head()
```

```python
fig = gmaps.figure()
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, max_intensity=10, point_radius=2)
fig.add_layer(heat_layer)
fig
```

```python
ideal_df = cities_df.loc[(cities_df['Temp']>70) & (cities_df['Temp']<80) & (cities_df['Wind Speed']<10) & (cities_df['Cloudiness']==0)]
ideal_df
```

```python
ideal_df['Hotel'] = ""
ideal_df['Hotel_Lat'] = ""
ideal_df['Hotel_Lng'] = ""
ideal_df
```

```python
params = {
    "distance": 5000,
    "rankby" : "distance",
    "type":"lodging",
    "key" : g_key
}

for index, row in ideal_df.iterrows():
    lat = row['Lat']
    lng = row['Lng']
    params["location"] = f"{lat}, {lng}"
    url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    response = requests.get(url, params=params).json()
    results = response['results']
    pprint(results) 
    try:
        ideal_df.loc[index, 'Hotel'] = results[0]['name']
        ideal_df.loc[index, 'Hotel_Lat'] = results[0]['geometry']['location']['lat']
        ideal_df.loc[index, 'Hotel_Lng'] = results[0]['geometry']['location']['lng']
    except(KeyError, IndexError):
        print("NA")
  

        
```

```python
ideal_df
```

```python
ideal_df_2 = ideal_df.dropna(how='any')
ideal_df_2
```

```python
hotel_name = ideal_df_2['Hotel'].tolist()
city_name= ideal_df_2['City'].tolist()
country_name = ideal_df_2['Country'].tolist()
```

```python
marker_locations = ideal_df_2[['Hotel_Lat', 'Hotel_Lng']]
fig = gmaps.figure()
markers = gmaps.marker_layer(marker_locations, info_box_content=[f"Hotel:{hotel}"for hotel in hotel_name],info_box_content=[f"City : {city}"for city in city_name], info_box_content=[f"Country :{country}"for country in country_name])
fig.add_layer(markers)
fig
```

```python
fig = gmaps.figure()
fig.add_layer(heat_layer)
fig.add_layer(markers)
fig
```
