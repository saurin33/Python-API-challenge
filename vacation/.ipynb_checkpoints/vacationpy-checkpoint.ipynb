```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import requests
import gmaps
from api_keys import g_key
from pprint import pprint
```

```python
locations = cities_df[['Lat', 'Lng']]
humidity = cities_df['Humidity']
```

```python
gmaps.configure(api_key = g_key)
```

```python
data = cities_df[['Lat', 'Lng', 'Humidity']]
data.head()
```

```python
fig = gmaps.figure()
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, max_intensity=10, point_radius=2)
fig.add_layer(heat_layer)
fig
```

```python
ideal_df = cities_df.loc[(cities_df['Temp']>70) & (cities_df['Temp']<80) & (cities_df['Wind Speed']<10) & (cities_df['Cloudiness']==0)]
ideal_df
```

```python
ideal_df['Hotel'] = ""
ideal_df
```

```python
params = {
    "distance": 5000,
    "rankby" : "distance",
    "type":"lodging",
    "key" : g_key
}

for index, row in ideal_df.iterrows():
    lat = row['Lat']
    lng = row['Lng']
    params["location"] = f"{lat}, {lng}"
    url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    response = requests.get(url, params=params).json()
    results = response['results']
    try:
        ideal_df[index, 'Hotel'] = results[0]['name']
    except(KeyError, IndexError):
        print("NA")
    

        
```
